# The Hydrogen Molecular Ion ($H_{2}^{+}$)

In this exercise you will apply **Molecular Orbital (MO) theory** to the simplest possible molecule, the hydrogen molecular ion ($H_{2}^{+}$).
You will:

1. Construct **bonding and antibonding molecular orbitals** from atomic orbitals.
2. Visualize these orbitals in real space.
3. Compute and analyze the **orbital energies as a function of internuclear distance**.

This system is simple enough to be treated semi-analytically, yet already illustrates the essential physics of chemical bonding.


### Physical System

The hydrogen molecular ion consists of:

* Two fixed nuclei $A$ and $B$ (Born–Oppenheimer approximation),
* One electron.

The internuclear separation is denoted by $R$.
The electron–nucleus distances are:

* $r_A$: distance to nucleus $A$,
* $r_B$: distance to nucleus $B$.



## Molecular Orbitals (LCAO Approximation)

We construct molecular orbitals as **linear combinations of atomic orbitals (LCAO)**:
$$
\psi_\pm = N_\pm \left( 1s_A \pm 1s_B \right)
$$

* $\psi_+$: bonding $1\sigma$ orbital
* $\psi_-$: antibonding $2\sigma$ orbital

The normalization constants are:
$$
N_\pm = \frac{1}{\sqrt{2 \pm 2S}}
$$
where $S$ is the **overlap integral**:
$$
S = \langle 1s_A | 1s_B \rangle
$$


## Overlap and Interaction Integrals

For hydrogenic 1s orbitals, the relevant integrals can be evaluated analytically.

#### Overlap Integral

$$
S(R) = \left( 1 + \frac{R}{a_0} + \frac{1}{3}\left(\frac{R}{a_0}\right)^2 \right)
e^{-R/a_0}
$$

#### Coulomb Integral

$$
j(R) = \frac{1}{R}
\left[ 1 - \left(1 + \frac{R}{a_0}\right) e^{-2R/a_0} \right]
$$

#### Exchange Integral

$$
k(R) = \frac{1}{a_0} \left(1 + \frac{R}{a_0}\right) e^{-R/a_0}
$$

where $ a_0 $ is the Bohr radius.

## Orbital Energies

The energy of the bonding orbital is:
$$
E_{1\sigma} = E_{\mathrm{H}_{1s}} + \frac{1}{R} - \frac{j + k}{1 + S}
$$

The energy of the antibonding orbital is:
$$
E_{2\sigma} = E_{\mathrm{H}_{1s}} + \frac{1}{R} - \frac{j - k}{1 - S}
$$

Here:

* $E_{\mathrm{H}_{1s}} = -\frac{1}{2}$ Hartree is the energy of the hydrogen 1s orbital,
* $\frac{1}{R}$ is the nuclear–nuclear repulsion term.
